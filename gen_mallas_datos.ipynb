```python
# Coordenadas de datos
x1 = 367384
y1 = 2531871
#x2 = 369289
#y2 = 2532839

# Incremento entre nodos
inc = 100
desfase_x = -150
desfase_y = -150
nodos_x = 23
nodos_y = 16

# Generar arreglos de coordenadas
lon = [x1+desfase_x]
lat = [y1+desfase_y]

for i in range(1,nodos_x,1):
    lon.append(lon[i-1]+inc)
for j in range(1,nodos_y,1):
    lat.append(lat[j-1]+inc)   

#print(lon), print(lat)

# Generar coordenadas de cada nodo
for i in lon:
    for j in lat:
        print(i,j)
```
